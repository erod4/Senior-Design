```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 44100         # Sampling frequency in Hz
N = 32            # Number of samples
t = np.arange(N) / fs  # Time vector

# Frequencies to include in the composite signal (Hz)
frequencies = [5, 10, 100, 500, 1000]

# Create the composite signal as the sum of sine waves
signal = sum(np.sin(2 * np.pi * f * t) for f in frequencies)
print(signal)
```

```python
# Compute the FFT of the signal
fft_signal = np.fft.fft(signal)
# Get corresponding frequency bins
freq_bins = np.fft.fftfreq(N, 1/fs)
```

```python
# For plotting, use only the positive frequencies
positive_indices = freq_bins >= 0
plt.figure(figsize=(10, 6))
plt.stem(freq_bins[positive_indices], np.abs(fft_signal)[positive_indices])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('FFT of Composite Sine Wave Signal')
plt.xlim(0, 5000)
plt.grid(True)
plt.show()
```
